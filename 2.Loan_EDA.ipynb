```python
# Before filling the missing value trim and convert into numeric dtype

df['emp_length']=df['emp_length'].str.replace(' years','',regex=False)
df['emp_length']=df['emp_length'].str.replace('10+years','10',regex=False)
df['emp_length']=df['emp_length'].str.replace('<1years','0.5',regex=False)
df['emp_length']=df['emp_length'].str.replace('n/ayears','0',regex=False)

# Convert into numeric dtype
df['emp_length']=pd.to_numeric(df['emp_length'],errors='coerce')
```

```python
df['emp_length'].median()
```

```python
#filling missing value with median
df['emp_length']=df['emp_length'].fillna(df['emp_length'].median())
```

```python
def util_bucket(x):
    if pd.isna(x):
        return 'unknown'
    elif  x<30:
        return 'low'
    elif x<70:
        return 'medium'
    else:
        return 'high'

    df['revol_util_bucket']=df['revol_util'].apply(util_bucket)
    df['revol_util_bucket'].value_counts()
   

```
